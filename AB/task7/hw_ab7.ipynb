Распишем алгоритм:


Наблюдательный подход:

1) Сэмплируем группы $Y_{.t}, Y_{.1}, X_{.t}, X_{.1}$, в начальный и конечный момент к эспериментальной группе добавляем эффект. Также в момент времени t добавляем эффект привыкания к эспериментальной группе

2) Рассчитываем оценку эффекта привыкания

3) Используем критерий Вальда для построения доверительного интервала и построения графика мощности.

Для исследования рассмотри, как минимум, три случая: 
- эффект привыкания есть;
- эффекта привыкания нет, как и прочих зависящих от времени эффектов, возникающих только в экспериментальной выборке;
- эффекта привыкания нет, но есть другой эффект, зависящий от времени, возникающий только в экспериментальной выборке, например, недельная сезонность (можно придумать что-то другое).

*Кажется, тут опять надо скопировать полотно разных функций, как-то их соединить и модифицировать. Например, генерация данных с эффектом была на последнем занятии.*

```python
num_users = 150000  # количество пользователей
period = 14  # длительность эксперимента

avg_usage_distr = sps.expon(scale=2000)  # распределение на avg_daily_usage
shows_up_distr = sps.bernoulli(p=0.5)  # распределение на shows_up
daily_variation_distr = sps.norm(
    loc=1, scale=4
)  # распределение на daily_variation
```

```python
def generate_raw_data(
    num_users, period, avg_usage_distr, shows_up_distr, daily_variation_distr
):
    """
    Генерирует данные описанным выше способом

    - num_users - количество пользователей
    - period - длительность эксперимента
    - avg_usage_distr - распределения на avg_daily_usage
    - shows_up_distr - распределение на shows_up
    - daily_variation_distr - распределение на daily_variation

    Возвращает
        data - DataFrame длины num_users * period
    """

    data = pd.DataFrame(range(num_users), columns=["user_id"])
    data["avg_daily_usage"] = avg_usage_distr.rvs(size=num_users)

    data["day"] = [range(period) for _ in range(num_users)]
    data = data.explode("day")
    data["day"] = data["day"].astype(int)

    data["shows_up"] = shows_up_distr.rvs(size=num_users * period)
    data["daily_variation"] = daily_variation_distr.rvs(size=num_users * period)

    data["daily_usage"] = (
        data["avg_daily_usage"] * data["shows_up"] * data["daily_variation"]
    )
    data.loc[data["daily_usage"] <= 0, "daily_usage"] = 0

    return data
```

```python
def get_exponential_model(mode):
    """
    Возвращает функцию для моделирования эффекта mode (novelty или primacy)
    """

    assert mode in ["novelty", "primacy"]

    def novelty_model(day, theta, gamma):
        return np.exp(-day * gamma) * theta

    def primacy_model(day, theta, gamma):
        return (1 - np.exp(-day * gamma)) * theta

    if mode == "novelty":
        return novelty_model
    else:
        return primacy_model


def split_groups(data, cnt_groups, target_column="group"):
    """
    Добавляет столбец target_column с номерами групп пользователей в data

    - data - DataFrame с данными
    - cnt_groups - количесто групп, на которые надо поделить пользователей
    - target_column - имя колонки с номером группы
    """

    users = data["user_id"].unique()
    np.random.shuffle(users)

    data[target_column] = 0

    for group in range(cnt_groups):
        group_idxs = data["user_id"].isin(users[group::cnt_groups])
        data.loc[group_idxs, target_column] = group


def add_effect(
    data,
    group,
    theta,
    gamma,
    scale,
    const_effect=0,
    mode="novelty",
    start_day=0,
    post_period_start=None,
    post_gamma=None,
    group_column="group",
    target_column="daily_usage",
):
    """
    Добавляет эффект в столбец column из распределения
        N(theta * exp(-gamma * t), scale^2) в случае mode == 'novelty'
        N(theta * (1 - exp(-gamma * t)), scale^2) в случае mode == 'primacy'

    - data - DataFrame с данными
    - group - группа пользователей, к которым надо добавить эффект
    - theta, gamma, scale - вышеописанные параметры эффекта
    - const_effect - постоянный эффект от влияния эксперимента
    - mode - определяет природу эффекта: novelty или primacy
    - start_day - номер дня старта эксперимента
    - post_period_start - номер дня старта post period при наличии этого периода
                          (для случая mode == primacy)
    - post_gamma - скорость отвыкания в post period (по умолчанию = gamma)
    - group_column - колонка с группами
    - target_column - колонка, к которой добавляется эффект
    """

    assert mode in ["novelty", "primacy"]
    assert not (mode == "novelty" and post_period_start is not None)
    if post_gamma is None:
        post_gamma = gamma

    # Выбираем индексы нужной группы
    idxs = (data[group_column] == group) & (data["day"] >= start_day)

    experiment_durations = data.loc[idxs, "day"] - start_day

    # Моделируем эффект
    model = get_exponential_model(mode=mode)
    mean = model(experiment_durations, theta, gamma) + const_effect

    # Если эффект primacy и передан момент старта постпериода,
    # то в нем привыкание заменяем отвыканием
    if mode == "primacy" and post_period_start is not None:
        # Вычисляем theta' - эффект на момент перед началом post-period
        last_le = model(post_period_start - start_day, theta, gamma)

        # Отвыкание моделируется той же функцией, что и эффект новизны
        unlearning_model = get_exponential_model(mode="novelty")

        # Заменяем привыкание отвыканием
        post_period_idxs = experiment_durations >= post_period_start - start_day
        days_from_start = experiment_durations[post_period_idxs] - (
            post_period_start - start_day
        )
        mean[post_period_idxs] = unlearning_model(
            days_from_start, last_le, post_gamma
        )

    # Генерируем шум и добавляем его с эффектом к данным
    noise = sps.norm(loc=0, scale=scale).rvs(len(mean))
    data.loc[idxs, target_column] += mean + noise


def prepare_data_obs_exp(
    data,
    theta,
    gamma,
    scale,
    const_effect=0,
    start_day=0,
    mode="novelty",
    group_column="group",
    target_column="daily_usage",
):
    """
    Готовит данные для наблюдательного подхода
    Делит пользователей на две группы, к одной в равных пропорциях добавляет
        эффект с параметрами из theta, gamma, scale

    - data - DataFrame с данными
    - theta, gamma, scale - вышеописанные параметры эффекта
                            (можно передавать листы параметров)
    - const_effect - постоянный эффект от влияния эксперимента
    - start_day - номер дня старта эксперимента
    - group_column - колонка с группами
    - target_column - колонка, к которой добавляется эффект
    """

    if not isinstance(theta, Iterable):
        theta = [theta]
        gamma = [gamma]
        scale = [scale]
    assert len(theta) == len(gamma) == len(scale)

    # Делим на группы: 2 выборки и каждую выборку еще на num_parts частей,
    # чтобы к каждой добавить эффект со своими параметрами
    num_parts = len(theta)
    split_groups(data, cnt_groups=2 * num_parts, target_column=group_column)
    for i, (t, g, s) in enumerate(zip(theta, gamma, scale)):
        # Добавляем эффект
        add_effect(
            data,
            group=i,
            theta=t,
            gamma=g,
            scale=s,
            const_effect=const_effect,
            mode=mode,
            start_day=start_day,
            group_column=group_column,
            target_column=target_column,
        )

    # Объединяем группы в выборки
    data.loc[data[group_column] < num_parts, group_column] = 0
    data.loc[data[group_column] >= num_parts, group_column] = 1


def prepare_data_ccd_exp(
    data,
    theta,
    gamma,
    scale,
    const_effect=0,
    start_day=0,
    period=period,
    mode="novelty",
    group_column="group",
    target_column="daily_usage",
):
    """
    Готовит данные для cookie-cookie-day подхода
    Делит пользователей на выборки в количестве равном длине эксперимента:
        i-ая выборка начинает получать эффект с i-ого дня

    - data - DataFrame с данными
    - theta, gamma, scale - вышеописанные параметры эффекта (list или число)
    - const_effect - постоянный эффект от влияния эксперимента
    - start_day - номер дня старта эксперимента
    - period - длительность эксперимента (задает количество выборок)
    - group_column - колонка с группами
    - target_column - колонка, к которой добавляется эффект
    """

    if not isinstance(theta, Iterable):
        theta = [theta]
        gamma = [gamma]
        scale = [scale]
    assert len(theta) == len(gamma) == len(scale)

    # Как и в предыдущей функции делим на period * num_parts частей:
    # period выборок и каждую выборку на num_parts частей
    num_parts = len(theta)
    split_groups(
        data, cnt_groups=period * num_parts, target_column=group_column
    )
    for sample_index in range(period):
        for i, (t, g, s) in enumerate(zip(theta, gamma, scale)):
            group = sample_index * num_parts + i

            # Добавляем эффект
            add_effect(
                data,
                group=group,
                theta=t,
                gamma=g,
                scale=s,
                const_effect=const_effect,
                mode=mode,
                start_day=sample_index + start_day,
                group_column=group_column,
                target_column=target_column,
            )

        # Объединяем в выборку
        idxs = (data[group_column] >= sample_index * num_parts) & (
            data[group_column] < (sample_index + 1) * num_parts
        )
        data.loc[idxs, group_column] = sample_index


def prepare_data_pp_exp(
    data,
    theta,
    gamma,
    scale,
    post_period_start,
    post_gamma=None,
    group_column="group",
    const_effect=0,
    start_day=0,
    target_column="daily_usage",
):
    """
    Готовит данные для post period подхода
    Делит пользователей на две группы, к одной в равных пропорциях добавляет
        эффект с параметрами из theta, gamma, scale

    - data - DataFrame с данными
    - theta, gamma, scale - вышеописанные параметры эффекта
                            (можно передавать листы параметров)
    - post_period_start - день старта post-period
    - post_gamma - gamma эффекта отвыкания, если отсутствует равна gamma
    - const_effect - постоянный эффект от влияния эксперимента
    - start_day - номер дня старта эксперимента
    - group_column - колонка с группами
    - target_column - колонка, к которой добавляется эффект
    """

    if not isinstance(theta, Iterable):
        theta = [theta]
        gamma = [gamma]
        scale = [scale]
        post_gamma = [post_gamma]
    if post_gamma is None:
        post_gamma = [None] * len(theta)

    assert len(theta) == len(gamma) == len(scale) == len(post_gamma)

    # Делим на 2 * num_parts частей
    num_parts = len(theta)
    split_groups(data, cnt_groups=2 * num_parts, target_column=group_column)
    for group, (t, g, s, p_g) in enumerate(
        zip(theta, gamma, scale, post_gamma)
    ):
        # Добавляем эффект
        add_effect(
            data,
            group=group,
            theta=t,
            gamma=g,
            scale=s,
            const_effect=const_effect,
            mode="primacy",
            start_day=start_day,
            post_period_start=post_period_start,
            post_gamma=p_g,
            group_column=group_column,
            target_column=target_column,
        )

    # Объединяем группы в выборки
    data.loc[data[group_column] < num_parts, group_column] = 0
    data.loc[data[group_column] >= num_parts, group_column] = 1


def plot_mean_groups(
    data,
    group_column,
    data_column,
    method_name="",
    start_day=None,
    post_period_start=None,
    group_names=None,
):
    """
    Рисует для каждой выборки средние значения по дням

    - data - DataFrame с данными
    - group_column - название колонки с группами
    - data_column - название колонки с данными
    - method_name - название метода для заголовка
    - start_day - номер дня старта эксперимента (опционально)
    - post_period_start - номер дня начала пост-периода (опционально)
    - group_names - названия групп (опционально)
    """

    if group_names is None:
        group_names = data[group_column].unique()

    # Вычисляем среднее для каждой выборки в каждый день
    means = data.groupby([group_column, "day"])[data_column].mean()

    plt.title(f"Среднее по дням в разных выборках\n{method_name}")

    for group_id, group_name in zip(
        sorted(data[group_column].unique()), group_names
    ):
        plt.plot(means.loc[group_id], label=str(group_name))

    if start_day is not None:
        plt.axvline(start_day - 0.5, label="Старт", color="r", ls="--")
    if post_period_start is not None:
        plt.axvline(
            post_period_start - 0.5,
            label="Начало пост-периода",
            color="black",
            ls="--",
        )

    plt.legend()


def plot_ccd_groups(
    data,
    group_column,
    data_column,
    method_name="",
    start_day=0,
    show_preperiod=False,
):
    """
    Рисует выборки в cookie-cookie-day подходе:
        одна линия - средние значения нулевой выборки по дням,
        другая линия - в день t = среднее значение выборки t

    - data - DataFrame с данными
    - group_column - название колонки с группами
    - data_column - название колонки с данными
    - method_name - название метода для заголовка
    - start_day - номер дня старта эксперимента
    - show_preperiod - рисовать ли значения нулевой выборки до start_day
    """

    # Вычисляем среднее для каждой выборки в каждый день
    means = (
        data.groupby([group_column, "day"])[data_column].mean().reset_index()
    )

    plt.title(f"Среднее по дням в разных выборках\n{method_name}")

    if not show_preperiod:
        means = means[means["day"] >= start_day]

    sns.lineplot(
        means[means[group_column] == 0],
        x="day",
        y=data_column,
        label="Долгосрочный эффект",
    )
    sns.lineplot(
        means[means["day"] - start_day == means[group_column]],
        x="day",
        y=data_column,
        label="Краткосрочный эффект\n(каждый день новая выборка)",
    )

    if start_day > 0 and show_preperiod:
        plt.axvline(start_day - 0.5, label="Старт", color="r", ls="--")


def plot_diffs(
    data, obs_cols=None, ccd_cols=None, period=period, theta=None, gamma=None
):
    """
    Строит графики оценки эффекта привыкания, приближающей его экспоненциальной
    модели и истинного среднего эффекта для случая эффекта новизны

    - data - DataFrame с данными
    - obs_cols - пары названий колонок с группой и значениями выборки для
                 наблюдательного подхода
    - ccd_cols - пары названий колонок с группой и значениями выборки для
                 cookie-cookie-day подхода
    - period - длина периода
    - theta, gamma - истинные параметры (листы или число)
    """

    days = np.arange(period)

    plt.figure(figsize=(10, 6))
    # Поскольку оцениваемый эффект стартует из нуля, тут модель такая же, как и для primacy эффекта
    model = get_exponential_model(mode="primacy")

    # Наблюдательный подход
    if obs_cols is not None:
        # Оцениваем эффект: вычисляем разность разностей
        obs_mean = data.groupby([obs_cols[0], "day"])[obs_cols[1]].mean()
        diff = obs_mean.loc[1] - obs_mean.loc[0]
        diff_in_diff = diff - diff[0]

        # Оцениваем параметры модели
        (est_theta, est_gamma), _ = curve_fit(
            model,
            xdata=days,
            ydata=diff_in_diff.values,
            bounds=([-1e4, 0], [1e4, 1]),
        )

        plt.plot(
            days, diff_in_diff, label=f"diff-in-diff", color="r", alpha=0.5
        )
        plt.plot(
            days,
            model(days, est_theta, est_gamma),
            color="r",
            ls="--",
            label=f"Экспоненциальная модель - diff-in-diff",
        )

    # Cookie-cookie-day подход
    if ccd_cols is not None:
        # Оцениваем эффект
        ccd_exp = (
            data[data[ccd_cols[0]] == 0].groupby(["day"])[ccd_cols[1]].mean()
        )
        ccd_control = (
            data[data[ccd_cols[0]] == data["day"]]
            .groupby(["day"])[ccd_cols[1]]
            .mean()
        )
        diff = ccd_control - ccd_exp

        # Оцениваем параметры модели
        (est_theta, est_gamma), _ = curve_fit(
            model, xdata=days, ydata=diff.values, bounds=([-1e4, 0], [1e4, 1])
        )

        plt.plot(days, diff, label=f"cookie-cookie-day", color="b", alpha=0.5)
        plt.plot(
            days,
            model(days, est_theta, est_gamma),
            color="b",
            ls="--",
            label=f"Экспоненциальная модель - cookie-cookie-day",
        )

    # Строим истинный средний эффект
    if theta is not None and gamma is not None:
        if not isinstance(theta, Iterable):
            theta = [theta]
            gamma = [gamma]

        effect = []
        for t, g in zip(theta, gamma):
            effect.append(model(days, t, g))

        plt.plot(
            days,
            np.mean(effect, axis=0),
            color="g",
            label=f"Истинный средний эффект",
        )

    plt.title("Эффект привыкания и его оценки экспоненциальными моделями")
    plt.xlabel("Время")
    plt.ylabel("Эффект")
    plt.legend()
```

```python
def calculate_delta_estimation(data, t):
    """Вычисление оценки привыкания для заданного момента времени t."""
    delta_group_1 = (
        data.loc[(data["group"] == 1) & (data["day"] == t), "daily_usage"].mean()
        - data.loc[(data["group"] == 1) & (data["day"] == 0), "daily_usage"].mean()
    )

    delta_group_0 = (
        data.loc[(data["group"] == 0) & (data["day"] == t), "daily_usage"].mean()
        - data.loc[(data["group"] == 0) & (data["day"] == 0), "daily_usage"].mean()
    )
    return delta_group_0 - delta_group_1


def calculate_sigma_estimation(data, t, sample_size):
    """Вычисление оценки дисперсии sigma для заданного момента времени t."""
    sigma2 = np.var(
        data.loc[(data["day"] == t) & (data["group"] == 0), ["daily_usage", "user_id"]]
        .groupby("user_id")
        .mean()
        .daily_usage,
        ddof=1,
    )
    p = calculate_correlation_for_common_users(data, t, 0)
    return np.sqrt(8 * (1 - p) * sigma2 / sample_size)


def calculate_obs_estimations(
    n_iter,
    sample_size,
    theta,
    gamma,
    scale,
    t,  # момент времени, когда производим оценку
    exp_period,
    mode="primacy",
    group_column="group",
    target_column="daily_usage",
    data_modification=None,  # функция для произволной модификации данных
    show_fig=False
):
    delta_estimations = []
    sigma_estimations = []

    for _ in tqdm(range(n_iter)):
        # Генерация и подготовка данных
        data = generate_raw_data(
            num_users=sample_size,
            period=exp_period,
            avg_usage_distr=sps.expon(scale=2000),
            shows_up_distr=sps.bernoulli(p=0.5),
            daily_variation_distr=sps.norm(loc=1, scale=0.1),
        )

        prepare_data_obs_exp(
            data,
            theta,
            gamma,
            scale,
            mode=mode,
            group_column=group_column,
            target_column=target_column,
        )

        if data_modification is not None:
            data_modification(data)

        # Добавление вычисленных оценок в списки
        delta_estimations.append(calculate_delta_estimation(data, t))
        sigma_estimations.append(calculate_sigma_estimation(data, t, sample_size))

        if show_fig:
            plot_mean_groups(
            data,
            group_column="group",
            data_column="daily_usage",
            method_name="эффект primacy, наблюдательный подход",
            start_day=0,
            group_names=["Эксперимент", "Контроль"],
        )

    return np.array(delta_estimations), np.array(sigma_estimations)


def calculate_correlation_for_common_users(data, t, group=0):
    common_users = set(
        data.loc[(data["day"] == t) & (data["group"] == group), "user_id"]
    ) & set(data.loc[(data["day"] == 0) & (data["group"] == group), "user_id"])

    usage_t = data.loc[
        (data["day"] == t)
        & (data["group"] == group)
        & (data["user_id"].isin(common_users)),
        ["user_id", "daily_usage"],
    ].sort_values("user_id")
    usage_0 = data.loc[
        (data["day"] == 0)
        & (data["group"] == group)
        & (data["user_id"].isin(common_users)),
        ["user_id", "daily_usage"],
    ].sort_values("user_id")

    usage_t_values = usage_t["daily_usage"].values
    usage_0_values = usage_0["daily_usage"].values

    if len(usage_t_values) > 1 and len(usage_0_values) > 1:
        p = np.corrcoef(usage_t_values, usage_0_values)[0, 1]
    else:
        p = np.nan

    return p
```

```python
def wald_rejections(theta_hats, sigma_hats, n,  show_fig=False, alpha=0.05,):
    z_stats = theta_hats  / sigma_hats

    # Двусторонний критерий (p-value для проверки H0: delta = 0)
    p_values = 2 * (1 - sps.norm.cdf(np.abs(z_stats)))

    # Отвергаем H0, если p-value < alpha
    rejections = p_values < alpha
    rejection_rate = np.mean(rejections)

    # Доверительный интервал для процента отвержений
    z_crit = sps.norm.ppf(1 - alpha / 2)
    margin = z_crit * np.sqrt((rejection_rate * (1 - rejection_rate)) / n)
    ci_lower = max(
        0, rejection_rate - margin
    )  # Гарантия, что не уходит в отрицательные значения
    ci_upper = min(1, rejection_rate + margin)  # Гарантия, что не превышает 1

    if show_fig:
        # Визуализация распределения z-статистик
        plt.figure(figsize=(7, 5))
        sns.histplot(z_stats, bins=30, kde=True, color="blue", alpha=0.6)
        plt.axvline(
            sps.norm.ppf(1 - alpha / 2),
            color="r",
            linestyle="dashed",
            label="Крит. значение (двусторонний)",
        )
        plt.axvline(-sps.norm.ppf(1 - alpha / 2), color="r", linestyle="dashed")
        plt.xlabel("z-статистика")
        plt.ylabel("Частота")
        plt.title("Распределение z-статистик")
        plt.legend()
        plt.show()

    return rejection_rate, ci_lower, ci_upper


rejection_percentage, ci_lower_percentage, ci_upper_percentage = wald_rejections(
    delta_estimations, sigma_estimations, n_iter, show_fig = True
)

# Вывод результатов
print(f"Процент отвержений: {rejection_percentage * 100:.2f}%")
print(
    f"Доверительный интервал на процент отвержений: ({ci_lower_percentage * 100:.2f}%, {ci_upper_percentage * 100:.2f}%)"
)
```

```python
def plot_rejection_percentage_ci(
    rejection_percentage, ci_lower_percentage, ci_upper_percentage
):
    fig, ax = plt.subplots(figsize=(10, 2))
    ax.fill_betweenx(
        [0.025, 0.075],
        ci_lower_percentage,
        ci_upper_percentage,
        color="green",
        alpha=0.6,
    )

    ax.axhline(0.05, color="grey")
    ax.axvline(
        x=0.05,
        color="red",
        linestyle="--",
    )
    ax.plot(rejection_percentage, 0.05, marker="*", color="r", markersize=15)

    ax.set_xlabel("Rejection Percentage")
    ax.set_ylabel("")
    ax.set_title(
        f"Rejection Interval: {ci_lower_percentage * 100:.2f}% - {ci_upper_percentage*100:.2f}%\nRejection Percentage: {rejection_percentage * 100:.2f}%"
    )

    # ax.set_xlim(0, 1)  # Устанавливаем пределы оси X от 0 до 10%
    ax.set_ylim(0, 0.1)  # Устанавливаем пределы оси Y для компактности

    ax.set_yticks([])

    plt.show()
```

```python
def estimate_power(n_iter, sample_size, thetas, gamma, scale, t, exp_period, alpha=0.05):
    """
    Оценивает мощность критерия Вальда при различных значениях эффекта \theta.
    """
    power_values = []
    
    for theta in thetas:
        delta_estimations, sigma_estimations = calculate_obs_estimations(
            n_iter, sample_size, theta, gamma, scale, t, exp_period
        )
        rejection_rate, _, _ = wald_rejections(delta_estimations, sigma_estimations, n_iter, alpha=alpha)
        power_values.append(rejection_rate)
    
    return np.array(power_values)

def plot_power_curve(theta_values, power_results_list, power_names):
    """
    Строит график мощности критерия для нескольких наборов мощностей.
    
    :param theta_values: Массив значений эффекта (θ).
    :param power_results_list: Список массивов мощностей для разных критериев.
    :param power_names: Список названий для каждого критерия.
    """
    plt.figure(figsize=(8, 5))
    
    # Для каждого массива мощностей и его названия строим график
    for power_results, name in zip(power_results_list, power_names):
        plt.plot(theta_values, power_results, marker='o', linestyle='-', label=name)
    
    plt.axhline(0.8, color='r', linestyle='--', label="Уровень значимости (80%)")
    plt.xlabel("Эффект (θ)")
    plt.ylabel("Мощность критерия")
    plt.title("График мощности критерия Вальда")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
def calculate_delta_estimation_ccd(data, t):
    """Вычисление оценки привыкания для метода CCD."""
    delta_group_0 = data.loc[
        (data["group"] == 0) & (data["day"] == t), "daily_usage"
    ].mean()
    delta_group_t = data.loc[
        (data["group"] == t) & (data["day"] == t), "daily_usage"
    ].mean()

    return delta_group_0 - delta_group_t


def calculate_sigma_estimation_ccd(data, t, sample_size):
    """Вычисление оценки дисперсии sigma для метода CCD."""
    sigma2 = np.var(
        data.loc[(data["day"] == t) & (data["group"] == 0), ["daily_usage", "user_id"]]
        .groupby("user_id")
        .mean()
        .daily_usage,
        ddof=1,
    )
    T = data["day"].max()
    return np.sqrt(2 * (T + 1) * sigma2 / sample_size)


def calculate_ccd_estimations(
    n_iter,
    sample_size,
    theta,
    gamma,
    scale,
    t,
    exp_period,
    mode="primacy",
    group_column="group",
    target_column="daily_usage",
    data_modification=None,
    show_fig=False,
):
    delta_estimations = []
    sigma_estimations = []

    for _ in tqdm(range(n_iter)):
        # Генерация и подготовка данных
        data = generate_raw_data(
            num_users=sample_size,
            period=exp_period,
            avg_usage_distr=sps.expon(scale=2000),
            shows_up_distr=sps.bernoulli(p=0.5),
            daily_variation_distr=sps.norm(loc=1, scale=0.1),
        )

        prepare_data_ccd_exp(
            data,
            theta,
            gamma,
            scale,
            mode=mode,
            group_column=group_column,
            target_column=target_column,
        )

        if data_modification is not None:
            data_modification(data)

        # Добавление вычисленных оценок в списки
        delta_estimations.append(calculate_delta_estimation_ccd(data, t))
        sigma_estimations.append(calculate_sigma_estimation_ccd(data, t, sample_size))

        if show_fig:
            plot_ccd_groups(
            test_data,
            group_column="group",
            data_column="daily_usage",
            method_name="эффект primacy, cookie-cookie-day подход",
            start_day=0,
            )

    return np.array(delta_estimations), np.array(sigma_estimations)
```

**1.** Предположим, что $\mathsf{H}_0$ верна. Оцени время до принятия решения в последовательном анализе с помощью семплирования и по теоретической формуле.

*Не забудь, что возможные ответы в последовательном анализе несколько отличаются от стандартного подхода, причем имеется симметрия. Если ты это понимаешь, то сможешь легко решить данный пункт на основе предыдущих.*

1) С помощью сэмплирования. Будем рассматривать параметры с семинара

```python
def calculate_stat(sample, logpdf_0, logpdf_1, cumulative=False):
    """
    Вычисление статистики критерия.

    sample - выборка
    logpdf_0 - функция вычисления логарифма плотности при H_0
    logpdf_1 - функция вычисления логарифма плотности при H_1
    cumulative - вернуть ли кумулятивную статистику или же только итоговую
    """

    z_stats = logpdf_1(sample) - logpdf_0(sample)
    if cumulative:
        return np.cumsum(z_stats, axis=-1)
    else:
        return np.sum(z_stats, axis=-1)
```

```python
def sequential_wald(samples, logpdf_0, logpdf_1, A, B, return_z=False):
    """
    Реализация последовательного критерия Вальда для исторических данных.

    sample - выборка
    logpdf_0 - функция вычисления логарифма плотности при H_0
    logpdf_1 - функция вычисления логарифма плотности при H_1
    A, B -- пороги
    return_z -- возвращать ли Z-статистики
    """

    z_stats = calculate_stat(samples, logpdf_0, logpdf_1, cumulative=True)

    # Текущий знак решения
    signs = (
        (z_stats >= np.log(B)).astype(int)
        - (z_stats <= np.log(A)).astype(int)
    )

    # Ищем первое ненулевое решение
    stop_time, decision = first_nonzero(signs)
    # Если такого нет, то мы не дошли до остановки, поставим максимальное время
    stop_time[decision == 0] = samples.shape[1]

    if return_z:
        return stop_time + 1, decision, z_stats
    else:
        return stop_time + 1, decision


def first_nonzero(values, axis=1):
    """
    Возвращает первый ненулевой элемент.
    Если таких нет, то возвращает 0.

    values -- двумерная матрица значений.
    axis -- ось, по которой производятся операции.
    """

    # Маска ненулевых элементов
    mask = values != 0
    # Индекс первого ненулевого элемента, если он есть.
    # Если его нет, то все нули, и возвращаем любой индекс, например, 0.
    first_nonzero_ind = np.where(mask.any(axis=axis), mask.argmax(axis=axis), 0)
    # Берем соответствующие элементы
    first_nonzero_values = values[range(values.shape[0]), first_nonzero_ind]

    return first_nonzero_ind, first_nonzero_values
```

```python
# Параметры гипотез
a0, a1 = 0, 0.25  # Средние для H0 и H1
scale = 1 # Дисперсия
alpha = 0.05       # Уровень значимости
beta = 0.2         # Вероятность ошибки 2-го рода

# Корректные пороги Вальда
A = beta / (1 - alpha)       # Порог для принятия H0
B = (1 - beta) / alpha       # Порог для отклонения H0

# Параметры симуляции
sample_count = 10000   # Количество выборок
sample_size = 150   # Максимальный размер выборки
size_range = np.arange(sample_size) + 1

# Генерация данных ПОД H0 (для оценки alpha)
samples = np.random.normal(loc=a0, size=(sample_count, sample_size))

z_stats = calculate_stat(
    samples, sps.norm(loc=a0).logpdf, sps.norm(loc=a1).logpdf, cumulative=True
)

sequential_wald_normal = lambda samples: sequential_wald(
    samples,
    sps.norm(loc=a0, scale=scale).logpdf,
    sps.norm(loc=a1, scale=scale).logpdf,
    A,
    B,
    return_z=True,
)

stop_time_values, decision = sequential_wald_normal(samples)[:2]
print(f"Среднее время остановки: {stop_time_values.mean():.2f}")
```

Теперь оценим через теоретическую формулу

Теоретическая формула.

$E_0 \tau  \approx \frac{\omega(\alpha, \beta)}{KL(P_0, P_1)}$, где 

$$\omega(\alpha, \beta) = \alpha \log \frac{\alpha}{\beta}  + (1  - \alpha) \log \frac{1 - \alpha}{\beta}$$

KL-дивергенция для двух нормальных распределений $\mathcal{N}(a_0, 1)$  и $\mathcal{N}(a_1, 1)$ задается так:

$$
KL(P_0, P_1) = \log\frac{1}{1} + \frac{1 + (a_0 - a_1)^2}{2} - \frac{1}{2}
$$

```python
omega = alpha * np.log(alpha / (1 - beta)) + (1 - alpha) * np.log((1 - alpha) / beta)
KL = (1 + (a0 - a1) ** 2) / 2 - 1 / 2
tau = omega / KL

print(f"Теоретическое среднее время до принятия решения: {tau:.2f} шагов")
```

**2.** Давай разберемся, можно ли совмещать два подхода &mdash; обычную процедуру проверки гипотез с критерием t-test и процедуру последовательного анализа с критерием Вальда. Для этого построим следующий критерий. Зафиксируем число $n$. До тех пор, пока размер выборки меньше $n$, будем пользоваться последовательным анализом и останавливать нашу процедуру в случае принятия одной из гипотез. Когда в выборке станет ровно $n$ объектов, а процедура последовательного анализа еще не завершилась, остановим ее и применим обычный t-test. Цель &mdash; оценить вероятность ошибки 1 рода у такой комбинации процедур.

*Указание.* Удобнее сразу сгенерировать выборки размера $n$, применить оба критерия независимо, а затем для наглядности построить <a href="https://miptstats.github.io/courses/python/10_pandas2.html#4.1-Функция-pd.crosstab">crosstab</a> по ответам каждого критерия, из нее уже можно получить ответ на вопрос задачи.

```python
def lm_ttest(x, y, cov_type="nonrobust", target_column="value"):
    """
    Реализация t-test для проверки равенства средних двух выборок с помощью
    сведения к t-test для проверки незначимости коэффициента в линейной регрессии.

    Параметры:
    x, y -- две выборки
    cov_type -- тип ковариационной матрицы, по умолчанию гомоскедастичная

    Возвращает:
    stat, pval -- статистика и p-value критерия
    """

    x["group"] = 0
    y["group"] = 1
    data = pd.concat([x, y])

    cov_kwds = {}
    if cov_type == "cluster":
        cov_kwds = {"groups": data["shop_id"]}

    model = sm.ols(f"{target_column} ~ group", data=data)
    results = model.fit(cov_type=cov_type, cov_kwds=cov_kwds)
    stat = results.t_test("group = 0").statistic
    pval = results.t_test("group = 0").pvalue
    return stat, pval

def logpdf_0(x):
    return sps.norm.logpdf(x, loc=a0, scale=1)

def logpdf_1(x):
    return sps.norm.logpdf(x, loc=a1, scale=1)
```

```python
def analyze_sequential_test(samples, n, logpdf_0, logpdf_1, A, B, a0, alpha):
    """
    Проводит анализ последовательного критерия Вальда с дооценкой t-тестом
    и рассчитывает доверительный интервал для ошибки 1 рода.

    Параметры:
    - samples: ndarray (M, L) - матрица выборок (M выборок длины L)
    - logpdf_0, logpdf_1: функции логарифмической плотности распределений H0 и H1
    - A, B: пороги для критерия Вальда
    - a0: среднее при H0 (для t-теста)
    - alpha: уровень значимости для t-теста

    Возвращает:
    - crosstab: pd.DataFrame - таблица сопряженности решений
    - alpha_estimate: float - оценка ошибки 1 рода
    - ci: tuple - 95% доверительный интервал ошибки 1 рода
    """
    # Запускаем последовательный критерий Вальда
    stop_times, decisions_seq = sequential_wald(
        samples[:, :n], logpdf_0, logpdf_1, A, B
    )

    # Массив итоговых решений
    final_decisions = decisions_seq.copy()

    # Выборки, для которых последовательный тест не дал решения (decision == 0)
    mask_no_decision = final_decisions == 0

    # Применяем t-тест к этим выборкам
    for i in np.where(mask_no_decision)[0]:
        sample = samples[i, :]
        t_stat, p_val = sps.ttest_1samp(sample, popmean=a0)
        final_decisions[i] = 1 if p_val < alpha else -1

    # Создаем массив решений с учетом t-теста
    t_decisions = decisions_seq.copy()
    t_decisions[mask_no_decision] = final_decisions[mask_no_decision]

    # Таблица сопряженности
    crosstab = pd.crosstab(
        pd.Series(decisions_seq, name="Sequential"),
        pd.Series(t_decisions, name="T-test / Final"),
    )

    # Оценка ошибки 1 рода
    k = (final_decisions == 1).sum()
    N = len(final_decisions)
    alpha_estimate = k / N

    # 95% доверительный интервал (метод Вальда)
    z = sps.norm.ppf(0.975)
    SE = np.sqrt(alpha_estimate * (1 - alpha_estimate) / N)
    ci_lower = max(0, alpha_estimate - z * SE)
    ci_upper = min(1, alpha_estimate + z * SE)

    ci = (ci_lower, ci_upper)

    return crosstab, alpha_estimate, ci
```

**3.** Теперь поймем, какая процедура мощнее. На занятии мы установили "стандартные" параметры. Возьми размер выборки $n$, который дает мощность 80% по формуле MDE для стандартного t-test. Промоделируй выборки при справедливости $\mathsf{H}_1$ и посчитай долю случаев, в которых процедура последовательного анализа принимает $\mathsf{H}_1$ до момента времени $n$ включительно. Сделай вывод относительно мощности исследуемых процедур.

*Не забывай указания предыдущих пунктов и про код с занятия.*

```python
# Параметры гипотез
a0, a1 = 0, 0.25  # Средние для H0 и H1
scale = 1  # Дисперсия
alpha = 0.05  # Уровень значимости
beta = 0.2  # Вероятность ошибки 2-го рода

# Корректные пороги Вальда
A = beta / (1 - alpha)  # Порог для принятия H0
B = (1 - beta) / alpha  # Порог для отклонения H0
```

```python
def sequential_analysis(samples, n, alpha=0.05):
    decisions = np.zeros(len(samples))  # Решения для каждой выборки
    for i, sample in enumerate(samples):
        for j in range(1, n+1):
            try:
                t_stat, p_val = sps.ttest_1samp(sample[:j], popmean=a0)
            except:
                break
            if p_val < alpha:
                decisions[i] = 1  # Отклоняем H0
                break
            if j == n:  # Если не отклонили до n
                decisions[i] = -1  # Принимаем H0
    return decisions

def compute_power(n, samples_H1):
    decisions_sequential = sequential_analysis(samples_H1, n, alpha)
    return (decisions_sequential == 1).mean()
```

**4.** На практике наблюдения могут приходить батчами по несколько объектов сразу. В таком случае нельзя устанавливать на них какой-либо порядок, применяя последовательный критерий Вальда по одному, необходимо добавлять сразу весь батч.

Исследуй, как меняется мощность процедуры последовательного анализа в зависимость от количества объектов в батче.

*Ты же понимаешь, что надо взять код из предыдущего пункта и указать один параметр?*

```python
def sequential_wald_batched(samples, batch_size, A, B, return_z=False):
    """
    Модифицированная версия с эквивалентной логикой, но другой реализацией
    """
    sample_count, n = samples.shape
    n_batches = n // batch_size
    trimmed_samples = samples[:, : n_batches * batch_size]

    # Переформатируем данные в батчи
    batched_samples = trimmed_samples.reshape(sample_count, n_batches, batch_size)

    # Вычисление LLR для каждого наблюдения
    llr = sps.norm(loc=a1, scale=scale).logpdf(batched_samples) - sps.norm(
        loc=a0, scale=scale
    ).logpdf(batched_samples)

    # Суммируем LLR внутри батчей и накапливаем
    cumulative_llr = np.cumsum(np.sum(llr, axis=2), axis=1)

    # Определяем пересечения порогов
    upper_mask = cumulative_llr >= np.log(B)
    lower_mask = cumulative_llr <= np.log(A)

    # Комбинированная маска решений
    decision_mask = upper_mask.astype(int) - lower_mask.astype(int)

    stop_points = np.argmax(decision_mask != 0, axis=1)

    ongoing = decision_mask[np.arange(sample_count), stop_points] == 0
    stop_points[ongoing] = n_batches - 1

    # Определяем финальные решения
    final_decisions = decision_mask[np.arange(sample_count), stop_points]

    if return_z:
        return stop_points + 1, final_decisions, cumulative_llr
    else:
        return stop_points + 1, final_decisions
```

```python
# Параметры
a0, a1 = 0, 0.25
scale = 1
alpha = 0.05
beta = 0.2

# Корректные пороги Вальда
A = beta/(1-alpha)  # ~0.2105
B = (1-beta)/alpha   # 16.0

# Эксперимент
batch_sizes = np.arange(1, n_t_test // 2, 10)
power_results = []

sample_size = 100000
samples = np.random.normal(loc=a1, scale=scale, size=(sample_size, n_t_test))

for bs in batch_sizes:
    stops, dec = sequential_wald_batched(samples, bs, A, B)
    power = (dec == 1).mean()
    power_results.append(power)
    print(f"Batch size: {bs:3d} | Power: {power:.3f}")

# Визуализация
plt.plot(batch_sizes, power_results, marker='o')
plt.xlabel('Размер батча')
plt.ylabel('Мощность')
plt.title('Зависимость мощности от размера батча\n(alpha=0.05, beta=0.2, MDE=0.8)')
plt.grid(True)
plt.show()
```

**5.** Наконец, давай вспомним, что последовательно приходящие данные не всегда являются выборкой, часто это временной ряд. Провалидируй процедуру последовательного анализа применительно к временным рядам.

*На занятии был код генерации временных рядов, используй его. И, конечно, код предыдущих пунктов задачи*.

```python
def generate_time_series(sample_size, time_period, a, scale):
    data = {
        'shop_id': np.repeat(np.arange(sample_size), time_period),
        'week_id': np.tile(np.arange(time_period), sample_size),
        'value': np.random.normal(a, scale, sample_size * time_period)
    }
    return pd.DataFrame(data)
```

```python
def sequential_wald_normal(series_array, mu0, mu1, sigma, A, B, return_z=False):
    n_shops, n_weeks = series_array.shape
    cum_llr = np.zeros((n_shops, n_weeks))  # Накопленная статистика LLR
    decision = np.zeros(n_shops)  # 0 — H0, 1 — H1
    z_stats = []  # Для хранения Z-статистик

    for i in range(n_shops):
        llr = np.zeros(n_weeks)
        for t in range(n_weeks):
            llr[t] = (series_array[i, t] - mu0) ** 2 / (2 * sigma**2) - (
                series_array[i, t] - mu1
            ) ** 2 / (2 * sigma**2)
            cum_llr[i, t] = np.sum(llr[: t + 1])  # Накопленная статистика LLR
            if cum_llr[i, t] >= np.log(B):
                decision[i] = 1
                break
            elif cum_llr[i, t] <= np.log(A):
                decision[i] = -1
                break
        z_stats.append(cum_llr[i])

    return np.array(decision), np.array(z_stats)


def plot_wald_analysis(cum_llr, weeks, n_show, A, B):

    plt.figure(figsize=(14, 8))
    for i in range(n_show):
        plt.plot(weeks, cum_llr[i], label=f"Магазин {i}", alpha=0.7, lw=3)

    # Отобразим горизонтальные пороги (в лог-пространстве)
    plt.hlines(
        np.log(B),
        1,
        len(weeks),
        colors="red",
        linestyles="--",
        lw=2,
        label="log(B) (принятие H1)",
    )
    plt.hlines(
        np.log(A),
        1,
        len(weeks),
        colors="blue",
        linestyles="--",
        lw=2,
        label="log(A) (принятие H0)",
    )
    plt.xlabel("Неделя")
    plt.ylabel("Накопленная статистика LLR")
    plt.title("Последовательный анализ для временных рядов")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
# Параметры

scale = 1500
alpha = 0.05
beta = 0.2
a0 = 5000
a1 = 5500

# Корректные пороги Вальда
A = beta / (1 - alpha)  # ~0.2105
B = (1 - beta) / alpha  # 16.0

sample_size = 1000  # Количество объектов
time_size = 20  # Длина временного ряда для каждого объекта
n_iter = 20_000  # Количество итераций семплирования
alpha = 0.05  # Уровень значимости
target_column = "value"  # Колонка с таргетом
```

```python
data = generate_time_series(sample_size, time_size, a=a1, scale=scale)

data_long = data.pivot(index="shop_id", columns="week_id", values="value")
values = data_long.values

decisions, cum_llr = sequential_wald_normal(values, a0, a1, scale, A, B, return_z=True)

power = np.mean(decisions == 1)
print(
    f"Процент магазинов, где H1 была принята: {power * 100:.2f}%"
)

n_show = 10  # число магазинов для визуализации
weeks = np.arange(1, time_size + 1)

plot_wald_analysis(cum_llr, weeks, n_show, A, B)
```
