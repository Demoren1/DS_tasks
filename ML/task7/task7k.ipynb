Рассмотрим одну из уже классических задач машинного обучения — предсказание кредитного рейтинга клиентов кредитной организации. В данной задаче вам нужно провести анализ данных, построить классификатор клиентов и отправить свои предсказания в Kaggle-соревнование.



Кредитный скоринг — система оценки кредитоспособности (кредитных рисков) потенциальных и действующих клиентов, в основу которой вкладываются статистические данные. Кредитный скоринг широко используется как крупными банками, микрофинансовыми организациями, так и в потребительском (магазинном) экспресс-кредитовании на небольшие суммы. Также возможно его использование в бизнесе сотовых операторов, страховых компаний и т. д. Скоринг заключается в присвоении баллов по заполнению некой анкеты, разработанной оценщиками кредитных рисков андеррайтерами. По результатам набранных баллов системой автоматически принимается решение об одобрении или отказе в выдаче кредита.



Описание данных:

* `client_id`: уникальный идентификатор клиента;

* `month`: месяц сбора информации;

* `age`: возраст клиента;

* `occupation`: вид деятельность;

* `annual_income`: годовой доход;

* `monthly_salary`: зарплата за текущий месяц;

* `number_credit_cards`: количество кредитных карт у данного клиента;

* `interest_rate`: процентная ставка по кредитной карте;

* `number_loan`: количество кредитов, взятых в банке;

* `delay_from_due_date`: среднее количество дней задержки выплат по кредиту;

* `number_delayed_payment`: количество платежей по кредиту, которые были выполнены с задержкой;

* `number_credit_inquiries`: количество запросов по кредиту;

* `unpaid_debt`: сумма долга;

* `credit_utilization_ratio`: коэффициент использования кредитных карт;

* `credit_history_age`: размер кредитной истории;

* `min_amount_payment`: была ли произведенная выплата минимально возможной;

* `monthy_amount_invested`: сумма инвестиций за месяц;

```python
def plot_histogram(
    data,
    column,
    bins=10,
    figsize=(10, 6),
    edgecolor="k",
    alpha=0.7,
    xlabel="Annual Income",
    ylabel="Frequency",
    title="Histogram",
    function_to_apply: callable = np.log,
):
    if function_to_apply is not None:
        data_filled = function_to_apply(data[column].dropna())
    else:
        data_filled = data[column].dropna()

    plt.figure(figsize=figsize)
    plt.hist(data_filled, bins=bins, edgecolor=edgecolor, alpha=alpha)

    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    # plt.show()


plot_histogram(
    data_train,
    title="Histogram of ",
    function_to_apply=lambda x: np.log(x + 2),
    column="number_bank_accounts",
)
```

```python
def apply_function_to_columns(dataframes, columns, function):
    for dataframe in dataframes:
        for column in columns:
            dataframe[column] = function(dataframe[column])


columns_to_transform = [
    "monthly_salary",
    "annual_income",
    "number_credit_cards",
    "number_delayed_payment",
    "age",
]

dataframes = [data_train, data_test]
apply_function_to_columns(dataframes, columns_to_transform, np.log)

columns_to_transform_special_first = [
    "delay_from_due_date",
]
apply_function_to_columns(
    dataframes, columns_to_transform_special_first, lambda x: np.log(x + 7)
)

columns_to_transform_special_second = [
    "monthy_amount_invested",
    "number_bank_accounts",
]
apply_function_to_columns(
    dataframes, columns_to_transform_special_second, lambda x: np.log(x + 2)
)
```

Далее я увидел, что в данных есть клиенты, которые есть в обоих датасетах. Я решил разделить их между train и test, сохраняя пропорцию.
Также оказалось, что присутствует корреляция между месяцами, поэтому я разделил данные по месяцам.

```python
# Разделяем данные по месяцам
train_months = ['Февраль', 'Апрель', 'Май', 'Август', 'Июнь']
test_months = ['Июль', 'Март', 'Сентябрь']

# Создаем train и test датасеты с непересекающимися месяцами
train_data = data_train[data_train['month'].isin(train_months)]
test_data = data_train[data_train['month'].isin(test_months)]

# Получаем уникальные client_id для train и test
train_clients = set(train_data['client_id'].unique())
test_clients = set(test_data['client_id'].unique())

# Находим общие client_id
common_clients = train_clients.intersection(test_clients)

# Определяем пропорцию распределения клиентов между train и test
train_ratio = len(train_clients) / (len(train_clients) + len(test_clients))

# Распределяем общих клиентов между train и test, сохраняя пропорцию
common_clients_list = list(common_clients)
np.random.shuffle(common_clients_list)
split_index = int(len(common_clients_list) * train_ratio)

train_common_clients = set(common_clients_list[:split_index])
test_common_clients = set(common_clients_list[split_index:])

# Обновляем train_data и test_data
train_data = pd.concat([train_data, test_data[test_data['client_id'].isin(train_common_clients)]])
test_data = pd.concat([test_data, train_data[train_data['client_id'].isin(test_common_clients)]])

# Удаляем дубликаты, если они появились
train_data.drop_duplicates(subset=['client_id', 'month'], keep='first', inplace=True)
test_data.drop_duplicates(subset=['client_id', 'month'], keep='first', inplace=True)

# Создаем X_train, y_train, X_test, y_test
X_train = train_data.drop(["credit_rating", "client_id", "month"], axis=1)
y_train = train_data["credit_rating"]
X_test = test_data.drop(["credit_rating", "client_id", "month"], axis=1)
y_test = test_data["credit_rating"]

# Подготавливаем X_kaggle
X_kaggle = data_test.drop(["client_id"], axis=1)
```

Найдем категориальные признаки:

```python
cat_features = [name for name, dtype in X_train.dtypes.items() if dtype == "object"]
cat_features
```

Преобразуем категориальные признаки в числа при помощи Mean Encoding

```python
for feature in cat_features:
    X_train[feature] = X_train[[feature]].fillna("nan")
    X_test[feature] = X_test[[feature]].fillna("nan")
    X_kaggle[feature] = X_kaggle[[feature]].fillna("nan")
```

Выберим самые важные признаки при помощи CatBoostClassifier.

```python
model = CatBoostClassifier(iterations=250, depth=6, learning_rate=0.1, verbose=0, l2_leaf_reg=10)

model.fit(X_train, y_train, cat_features=cat_features)  

importances = model.get_feature_importance()

feature_importance = pd.Series(importances, index=X_train.columns).sort_values(ascending=False)

print(feature_importance)
```

В качестве признаков выберем 10 самых важных признаков

```python
X_train = X_train[feature_importance.head(10).index]
X_test = X_test[feature_importance.head(10).index]
X_kaggle = X_kaggle[feature_importance.head(10).index]

# best_importances = [
#     "age",
#     "annual_income",
#     "monthly_salary",
#     "number_bank_accounts",
#     "number_credit_cards",
#     "unpaid_debt",
#     "interest_rate",
#     "delay_from_due_date",
#     "monthy_amount_invested",
#     "monthly_balance",
# ]

# X_train = X_train[best_importances]
# X_test = X_test[best_importances]
# X_kaggle = X_kaggle[best_importances]

cat_features = [name for name, dtype in X_train.dtypes.items() if dtype == "object"]
X_train.head()
```

```python
params = {'n_estimators': 500, 'learning_rate': 0.2, 'l2_leaf_reg': 3, 'depth': 8}

model = CatBoostClassifier(
    eval_metric="Accuracy", use_best_model=True, random_state=0, verbose=100, **params
)

model.fit(X_train, y_train, eval_set=(X_test, y_test), cat_features=cat_features)

y_pred = model.predict(X_test)

result = accuracy_score(y_true=y_test, y_pred=y_pred)

print(f"Accuracy: {result:.3f}")
```

```python
def show_prediction(model, X_test, y_test):
    pred = model.predict(X_test)
    print(f"Accuracy: {100 * accuracy_score(y_test, pred):.2f}")
    return 100 * accuracy_score(y_test, pred)


def save_predictions_in_csv(model, data_to_predict, name="prediciton"):
    pred = model.predict(data_to_predict).ravel()
    tmp_preds = pd.DataFrame({"index": data_test.index, "credit_rating": pred})
    tmp_preds.set_index("index", inplace=True)
    tmp_preds.to_csv(name + ".csv", index=True)
```
