Для каждого случая из перечисленных выше постройте график плотности сопряженного распределения (на одной фигуре).

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sps
import seaborn as sns
```

Мы знаем, что к бернуллевскому распределению является бета распределение

Подберем параметры $\alpha$ и $\beta$ так, чтобы они соотвествовали предполагаемому распределению

```python
x = np.linspace(0, 1, 1000)

experiments_names = [
    "Монета скорее нечестная, перевес неизвестен",
    "Монета скорее нечестная, перевес неизвестен",
    "Монета скорее нечестная, перевес в сторону герба",
    "Монета скорее честная, либо с небольшим перекосом вправо",
    "Ничего не известно",
]
alpha_n = [20, 0.7, 10, 20, 1]
beta_n = [20,  0.7,  2, 13.6, 1]

plt.figure(figsize=(14, 8))

for name, alpha, beta in zip(experiments_names, alpha_n, beta_n):
    y = sps.beta(alpha, beta).pdf(x)
    plt.plot(x, y, label=name, lw=2)

plt.legend()
plt.xlabel('Вероятность успеха (p)', fontsize=14)
plt.ylabel('Плотность', fontsize=14)
plt.title('Бета-распределения для различных априорных предположений', fontsize=16)
plt.grid(alpha=0.3)

plt.show()
```

Реализуйте функцию, вычисляющую параметры априорного распределения:

Если $q(t) \sim \text{Beta}(\alpha, \beta)$  то $q(t|X) \sim \text{Beta}(\alpha + \sum X_i, \beta + n - \sum X_i)$

```python
def bern_posterior_params(sample, a, b):
    '''Находит параметры апостериорного распределения 
    
    :param sample: выборка
    :param a: параметр alpha априорного распределения
    :param b: параметр beta априорного распределения
    
    :returns: параметры апостериорного
    '''
    return a + np.sum(sample), b + len(sample) - np.sum(sample)
```

**1)** В схеме испытаний Бернулли из задачи 4 рассмотрим онлайн-подход: наблюдения поступают батчами, в качестве априорного используется предыдущее апостериорное. Сначала реализуйте функцию генерации батчей:

```python
def get_sample_bern(theta=0.5, batch_size=5):
    '''
    :returns: батч размера <batch_size> из распределения Bern(<theta>)
    '''
    return np.random.binomial(1, theta, batch_size)
```

Допишите функцию, имитирующую онлайн подход:

```python
test = [(1, 2), (3, 4)]
```

```python
def bern_posterior_params_batch(batch_count, a_prior, b_prior, theta=0.5, batch_size=5):
    '''
    :param batch_count: количество батчей в выборке
    :param a_prior, b_prior: параметры априорного распределения (до получения выборки)
    :param theta: истинный параметр распределения бернулли
    :param batch_size: размер батча. Размер выборки в таком случае <batch_size> * <batch_count>
    
    :returns: набор апостериорных параметров, полученных после каждого нового батча
    '''
    posterior_params = [(a_prior, b_prior)]

    for _ in range(batch_count):
        batch = get_sample_bern(theta, batch_size)
        a_poster, b_poster = bern_posterior_params(batch, *posterior_params[-1])
        posterior_params.append((a_poster, b_poster))

    return np.array(posterior_params)
```

Ниже приведена реализация вспомогательной функции

```python
def draw_posteriori(grid, distr_class, post_params, xlim=None, true_p=0.5):
    '''Рисует серию графиков апостериорных плотностей.
    
    :param grid: сетка для построения графика
    :param distr_class: класс распределений из scipy.stats
    :param post_params: параметры апостериорных распределений 
        shape=(размер выборки, кол-во параметров)
    '''
        
    size = post_params.shape[0] - 1
    
    plt.figure(figsize=(12, 7))
    for n in range(size+1):
        plt.plot(grid, 
                 distr_class(post_params[n]).pdf(grid) \
                     if np.isscalar(post_params[n]) \
                     else distr_class(*post_params[n]).pdf(grid), 
                 label='n={}: {}'.format(n, np.round(post_params[n], 3)), 
                 lw=2.5,
                 color=(1-n/size, n/size, 0))
    plt.title('Апостериорное распределение в зависимости от номера батча')
    plt.axvline(true_p, color="red", linestyle='-.', label=f"Истинное значение: p = {true_p:.2f}")
    plt.legend(fontsize=12)
    plt.xlim(xlim)
    plt.show()
```

**2)** Проведите аналогичное исследование для выборки из распределения $Exp(\theta)$

С лекции знаем, что гамма распределение является сопряженным к экспоненциальному

Причем, если $q(t) \sim \Gamma (\alpha, \beta + 1)$, тогда $q(t|X) \sim \Gamma (\alpha + \sum X_i, \beta + n + 1)$

```python
def get_sample_exp(theta=1, batch_size=5):
    '''
    :returns: батч размера <batch_size> из распределения Exp(<theta>)
    '''
    return np.random.exponential(theta, batch_size)

def exp_posterior_params(batch_count, a_prior, b_prior, theta=1, batch_size=5):
    '''
    :param batch_count: количество батчей в выборке
    :param a_prior, b_prior: параметры априорного распределения (для гамма-распределения)
    :param theta: истинный параметр экспоненциального распределения
    :param batch_size: размер батча. Размер выборки в таком случае <batch_size> * <batch_count>
    
    :returns: набор апостериорных параметров, полученных после каждого нового батча
    '''
    
    posterior_params = [(a_prior, b_prior)]  # Начальные априорные параметры

    for _ in range(batch_count):
        # Получаем выборку из экспоненциального распределения
        batch = get_sample_exp(theta, batch_size)
        
        # Обновляем параметры апостериорного распределения
        a_poster = posterior_params[-1][0] + batch_size  # Обновляем параметр a (добавляем batch_size)
        b_poster = posterior_params[-1][1] + np.sum(batch)  # Обновляем параметр b (добавляем сумму выборки)

        # Добавляем новые параметры в список
        posterior_params.append((a_poster, b_poster))

    return np.array(posterior_params)

def draw_posteriori_gamma(grid, post_params, xlim=None, true_p=0.5):
    '''Рисует серию графиков апостериорных плотностей для гамма-распределения.
    
    :param grid: сетка для построения графика
    :param distr_class: класс распределений из scipy.stats, в данном случае gamma
    :param post_params: параметры апостериорных распределений 
        shape=(размер выборки, 2) — так как для гамма распределения нужны два параметра
    :param xlim: пределы по оси x для графика (опционально)
    :param true_p: истинное значение (опционально, для вертикальной линии на графике)
    '''
        
    size = post_params.shape[0] - 1
    
    plt.figure(figsize=(12, 7))
    for n in range(size + 1):
        # Извлекаем параметры alpha и beta
        alpha, beta = post_params[n]
        
        # Строим плотность для гамма-распределения
        plt.plot(grid, 
                 sps.gamma(alpha, scale=1/beta).pdf(grid), 
                 label='n={}: α={}, β={}'.format(n, np.round(alpha, 3), np.round(beta, 3)), 
                 lw=2.5,
                 color=(1-n/size, n/size, 0))  # Цвет графика меняется с каждым шагом
        
    plt.title('Апостериорное распределение для гамма-распределения в зависимости от номера батча')
    plt.axvline(true_p, color="red", linestyle='-.', label=f"Истинное значение: p = {true_p:.2f}")
    plt.legend(fontsize=12)
    plt.xlim(xlim)
    plt.show()
```

1) Априорные параметры соответствуют реальности

```python
grid = np.linspace(0, 10, 1000)
theta = 0.5
a_prior = 5
b_prior = 2.5
batch_count = 10
batch_size = 10

posterior_params = exp_posterior_params(batch_count, a_prior, b_prior, theta, batch_size)
draw_posteriori_gamma(grid, posterior_params, (0, 5), 1/theta)
```

2) Возьмем неподходящие параметры.

```python
grid = np.linspace(0, 10, 1000)
theta = 0.5
a_prior = 1
b_prior = 1
batch_count = 10
batch_size = 10

posterior_params = exp_posterior_params(batch_count, a_prior, b_prior, theta, batch_size)
draw_posteriori_gamma(grid, posterior_params, (0, 5), 1/theta)
```

После получения очередной партии наблюдений вычисляется апостериорное распределение параметров, используя в качестве априорного распределения предыдущее апостериорное распределение.
```python
from scipy.stats import gaussian_kde

class EmpiricalCDF:
    def __init__(self, dim):
        self.dim = dim
        self.data = np.empty((0, dim))  # Храним точки
        self.n = 0  # Количество точек

    def update(self, new_points):
        """Обновление выборки"""
        new_points = np.atleast_2d(new_points)
        self.data = np.vstack([self.data, new_points])
        self.n += len(new_points)

    def evaluate(self, x):
        """Оценка CDF в точке x"""
        return np.mean(np.all(self.data <= x, axis=1))

    def kde_estimate(self):
        """Оценка плотности с помощью KDE"""
        if self.n < 2:
            raise ValueError("Недостаточно данных")
        kde = gaussian_kde(self.data.T)
        return kde
```

```python
def get_batch_norm(theta_1, theta_2, batch_size):
    """Генерация батча из нормального распределения"""
    return np.random.normal(theta_1, np.sqrt(1 / theta_2), size=batch_size)

def norm_posterior_params(batch_count, a_prior, b_prior, theta_1=0, theta_2=1, batch_size=5):
    """
    Оценка параметров нормального распределения через последовательное обновление.

    :param batch_count: количество батчей в выборке
    :param a_prior: априорное среднее
    :param b_prior: априорный precision (1/дисперсия)
    :param theta_1: истинное значение среднего
    :param theta_2: истинное значение precision (1/дисперсия)
    :param batch_size: размер батча
    
    :returns: набор апостериорных параметров и эмпирическое распределение
    """
    
    posterior_params = [(a_prior, b_prior)]  # Начальные априорные параметры
    posterior_distr = EmpiricalCDF(dim=2)
    posterior_distr.update(np.array([posterior_params[-1]]))

    for _ in range(batch_count):
        # Получаем выборку из нормального распределения
        batch = get_batch_norm(theta_1, theta_2, batch_size)
        
        # Оценка параметров нормального распределения
        a_poster = np.mean(batch)
        b_poster = 1 / np.var(batch)  # precision = 1 / variance

        # Комбинируем с предыдущими параметрами
        a_poster = np.mean([a_poster, a_prior])
        b_poster = np.mean([b_poster, b_prior])
        a_prior, b_prior = posterior_params[-1]

        posterior_params.append((a_poster, b_poster))
        posterior_distr.update(np.array([posterior_params[-1]]))

    return np.array(posterior_params), posterior_distr
```

```python
theta_1 = 0
theta_2 = 1
a_prior = -10
b_prior = 0.01
batch_count = 500
batch_size = 5

posterior_params, posterior_distr = norm_posterior_params(batch_count, a_prior, b_prior, theta_1, theta_2, batch_size)   
```
