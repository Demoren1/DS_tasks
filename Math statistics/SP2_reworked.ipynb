## Задача 1. 


**1.** Пусть $X_1, ..., X_n$ &mdash; выборка из распределения $U[0, \theta]$. Рассмотрим оценки $X_{(n)}, \frac{n+1}{n}X_{(n)}, 2\overline{X}$ параметра $\theta$.

Какие из этих оценок являются несмещенными?

**Ответ:** Несмещенными являются $\frac{n+1}{n}X_{(n)}, 2\overline{X}$.(Пояснения смотреть в №1 теор. д/з.)

Проведите практическое исследование несмещенности перечисленных оценок аналогично ислледованию несмещенности выборочной дисперсии в предыдущих заданиях. Разбор этой задачи для выборочной дисперсии был на занятии.

**Решение:**

Без потери общности, примем $\theta = 1$ 

```python
def estimate_n_ord_stat(X):
    n = X.shape[1] - 1
    return np.apply_along_axis(lambda x: np.partition(x, n)[n], axis=1, arr=X)


def estimate_correctted_n_ord_stat(X):
    n = X.shape[1] - 1
    return np.apply_along_axis(
        lambda x: np.partition(x, n)[n] * (n + 1) / n, axis=1, arr=X
    )


def estimate_2mean(X):
    return np.apply_along_axis(lambda x: 2 * np.mean(x), axis=1, arr=X)
```

```python
theta = 1
n = 10
X_s = sps.norm.rvs(size=(5, 3))

estimators = [
    (estimate_n_ord_stat, r"$X_{(n)}$", "red", 0),
    (estimate_correctted_n_ord_stat, r"$\frac{n+1}{n}X_{(n)}$", "green", 0.1),
    (estimate_2mean, r"$2\overline{X}$", "blue", 0.2)
]
```

```python
def est_plot(distribution, estimators, sample_size, sample_count):  
    '''
    Построение графика разброса реализаций оценок и их средних значений.
    
    distribution -- распределение формата scipy.stats
    estimators -- список оценок и параметров для отрисовки графиков
    sample_size -- размер выборок
    sample_count -- количество генерируемых выборок
    '''    

    X = distribution.rvs(size=(sample_count, sample_size))
    
    plt.figure(figsize=(18, 0.7 * len(estimators)))
    for estimator, name, color, y in estimators:
        E = estimator(X)
        # print("E", E.shape)
        # print("y", (np.zeros(sample_count) + y).shape)
        plt.scatter(E, np.zeros(sample_count) + y, alpha=0.1,
                    s=100, color=color, label=name)
        plt.scatter(E.mean(), y, marker='*', s=300,
                    color='w', edgecolors='black')

    plt.vlines(1, -1, 1, color='r')
    plt.title('Размер выборки = %d' % sample_size)
    plt.yticks([])
    plt.legend()
    plt.xlim((0.6, 1.4))
    plt.ylim((-0.1, 0.1 * len(estimators)))
```

```python
sample_size_list = (10, 100, 500)  # размеры выборок
sample_count = 500  # количество экспериментов

for sample_size in sample_size_list:
    est_plot(sps.uniform, estimators, sample_size, sample_count)
```

## Задача 2.

Пусть $X_1, \ldots, X_n$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$, причем $a$ не известно. Визуализируйте свойства состоятельности и асимптотической нормальности для оценки $S^2 = \overline{X^2} - \overline{X}^2$ дисперсии $\sigma^2$. 

Примеры кода можно посмотреть в материалах лекций. Обязательно на графике оценок визуализируйте границы, которые задает свойство асимптотической нормальности.

**Решение:**

```python
a = 0
sigma2 = 1
sigma = np.sqrt(sigma2)
size = 300  # размер выборок
samples_count = 500  # количество выборок
n_range = (np.arange(size) + 1)  # размеры подвыборок

samples = sps.norm(a, sigma).rvs(size=(samples_count, size))
```

```python
def estimate_s2(X):
    ''' Смещенная оценка дисперсии '''
    return X.var(axis=-1)

def cumulative_variance_matrix(samples):
    n_rows, n_cols = samples.shape
    cum_vars = np.zeros((n_rows, n_cols))
    
    for i in range(1, n_cols + 1):
        cum_var = estimate_s2(samples[:, :i])
        cum_vars[:, i-1] = cum_var
    
    return cum_vars
```

```python
estimation = cumulative_variance_matrix(samples)
```

Границы можно построить использую правило 2 $\sigma$. 

```python
plt.figure(figsize=(15, 7))
for i in range(samples_count):
    plt.plot(np.arange(size) + 1, estimation[i], color="blue", alpha=0.05)
plt.plot(n_range, sigma2 + 3 * sigma2 / np.sqrt(n_range), color="red", lw=3)
plt.plot(n_range, sigma2 - 3 * sigma2 / np.sqrt(n_range), color="red", lw=3)
plt.title("Поведение оценки для разных реализаций")
plt.xlabel("Размер выборки")
# plt.ylim((0, 5))
plt.xlim((0, size))
```

<b><font color="#00b565" size="5">Исправил код</font></b>

<b><font color="#00b565" size="3">Замечание:</font></b> <i> Нет визуализации. Комм. Нет визуализации распределения для статистики $Z_n$.</i>

Визуализируем статистику $Z_n = \sqrt{n} \cdot \frac{\hat{\theta} - \theta}{\sqrt{2}\sigma^2}$, которая по ЦПТ должна сходится к стандартному нормальному распределению.

```python
Zn = np.sqrt(n_range) * (estimation - theta) / (theta)

plt.figure(figsize=(12, 8))
grid = np.linspace(-3, 3, 1000)

plt.hist(Zn[:, -1], bins=25, density=True)
plt.plot(grid, sps.norm.pdf(grid), lw=5, label="$\\mathcal{N}(0, 1)$")
plt.legend()
plt.title('Аппроксимация $Z_n$')

plt.show()
```

Как видим, гистограмма $Z_n$ действительно отвечает нормальному распределению.

<b><font color="#00b565" size="5">===== Конец исправлений =====</font></b>

**Вывод:** 
1. С ростом размера выборки траектории оценок сходятся к истинному значению параметра, об этом говорит свойство состоятельности.

2. Свойство асимптотической нормальности говорит о том, что с вероятностью около 0.95 (в силу уже упомянутого правила $2\sigma$) при больших размерах выборки значение оценки отклоняется от истинного значения параметра не более чем на $\frac{2\sigma}{\sqrt{n}}$. Это также можно увидеть на графике.

## Задача 3.

Пусть $X = (X_1, ..., X_n)$ &mdash; выборка из распределения $\mathcal{N}(a, \sigma^2)$, причем $a$ не известно. Визуализируйте асимптотический доверительный интервал Вальда для параметра дисперсии при росте размера выборки. 

Для этого сгенерируйте выборку $X_1, ... X_{N}, N = 100$ и постройте график доверительных интервалов уровня доверия $\alpha=0.95$, вычисленных для всех подвыборок размера $n$ вида $X_1, ... X_n$, $1 \le n \le 100$, используя написанную выше функцию. На график также нужно нанести точки выборки.

Чтобы не плодить код, напишите следующую функцию (см. ниже). При выборе стиля графика помните, что если изображаются лишь точки и линии, то лучше использовать серый фон, а если присутствуют закрашенные области, то предпочтительнее белый. Графики первого типа еще называют "легкими", а второго — "тяжелыми".

Для вычисления квантилей у каждого распределения из `scipy.stats` используйте функцию `ppf`.

**Решение:**

```python
def draw_confidence_interval(
    left,
    right,
    estimations=None,
    sample=None,
    ylim=(None, None),
    color_estimation="#FF3300",
    color_interval=["#00CC66"],
    color_sample="#0066FF",
    label_estimation=[" "],
    dots_flag=False,
    title_name=None,
    true_values=None,
):
    """
    Рисует доверительный интервал и оценку в зависимости от размера выборки.

    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param color_estimation: цвет оценки
    :param color_interval: цвет интервала
    :param color_sample: цвет выборки
    :param label_estimation: подпись для оценки
    """
    plt.figure(figsize=(12, 8))

    for tmp_left, tmp_right, tmp_color_interval, tmp_label_estimation,tmp_estimation in zip(
        left, right, color_interval, label_estimation, estimations
    ):
        size = estimation.shape[0]
        x = np.arange(size) + 1

        plt.plot(
            x,
            tmp_estimation,
            marker=".",
            color=tmp_color_interval,
            label=r"Оценка " + tmp_label_estimation,
        )

        plt.plot(x, tmp_right, ls="-", color=tmp_color_interval)
        plt.plot(x, tmp_left, ls="-", color=tmp_color_interval)
        plt.fill_between(
            x,
            tmp_left,
            tmp_right,
            color=tmp_color_interval,
            alpha=0.2,
            label=tmp_label_estimation,
        )

        if dots_flag:
            plt.plot(
                x,
                sample,
                alpha=0.55,
                ls=" ",
                marker="o",
                color="orange",
                label="Выборка",
            )

    if true_values is not None:
        colors = ["r", "b", "g"]
        for i, true_val in enumerate(true_values):
            plt.axhline(y=true_val, color=colors[i], linestyle='--', label=r"Истинное значение")

    plt.xlim((10, size + 1))
    plt.ylim(ylim)

    plt.xlabel("Размер выборки")
    plt.ylabel("Значение дисперсии")
    plt.legend()
    plt.title(title_name)
    plt.grid(True)
    plt.show()
```

```python
def draw_confidence_interval(
    left,
    right,
    estimations=None,
    sample=None,
    ylim=(None, None),
    color_estimation="#FF3300",
    color_interval=["#00CC66"],
    color_sample="#0066FF",
    label_estimation=[" "],
    dots_flag=False,
    title_name=None,
    true_values=None,
):
    """
    Рисует доверительный интервал и оценку в зависимости от размера выборки.

    :param left: левые границы интервалов (в зависимости от n)
    :param right: правые границы интервалов (в зависимости от n)
    :param estimation: оценки (в зависимости от n)
    :param sample: выборка
    :param ylim: ограничение вертикальной оси
    :param color_estimation: цвет оценки
    :param color_interval: цвет интервала
    :param color_sample: цвет выборки
    :param label_estimation: подпись для оценки
    """
    plt.figure(figsize=(12, 8))

    for tmp_left, tmp_right, tmp_color_interval, tmp_label_estimation,tmp_estimation in zip(
        left, right, color_interval, label_estimation, estimations
    ):
        size = estimation.shape[0]
        x = np.arange(size) + 1

        plt.plot(
            x,
            tmp_estimation,
            marker=".",
            color=tmp_color_interval,
            label=r"Оценка " + tmp_label_estimation,
        )

        plt.plot(x, tmp_right, ls="-", color=tmp_color_interval)
        plt.plot(x, tmp_left, ls="-", color=tmp_color_interval)
        plt.fill_between(
            x,
            tmp_left,
            tmp_right,
            color=tmp_color_interval,
            alpha=0.2,
            label=tmp_label_estimation,
        )

        if dots_flag:
            plt.plot(
                x,
                sample,
                alpha=0.55,
                ls=" ",
                marker="o",
                color="orange",
                label="Выборка",
            )

    if true_values is not None:
        colors = ["r", "b", "g"]
        for i, true_val in enumerate(true_values):
            plt.axhline(y=true_val, color=colors[i], linestyle='--', label=r"Истинное значение")

    plt.xlim((10, size + 1))
    plt.ylim(ylim)

    plt.xlabel("Размер выборки")
    plt.ylabel("Значение дисперсии")
    plt.legend()
    plt.title(title_name)
    plt.grid(True)
    plt.show()
```

Мы помним, что $S^2$ является а.н.о. для $\sigma^2$.


<b><font color="#00b565" size="5">Добавил формулы для интервалов </font></b>

<b><font color="#00b565" size="3">Замечание:</font></b> <i>Нет никаких формул для интервалов. Комм. Не выписаны явно формулы для интервалов, они есть только в коде.</i>

Из теоретической домашней работы нам известно, что а.д.и. для $\sigma^2$ имеет вид

$$
\left(S^2 - z_{\frac{1 + \alpha}{2}} \cdot \frac{\sqrt{2}S^2}{\sqrt{n}}, S^2 + z_{\frac{1 + \alpha}{2}} \cdot \frac{\sqrt{2}S^2}{\sqrt{n}}\right)
$$

<b><font color="#00b565" size="5">===== Конец исправлений =====</font></b>

```python
N = 100
alpha = 0.95
a = 0
sigma = 1

z = sps.norm.ppf((1 + alpha) / 2)

samples = sps.norm(a, sigma).rvs(size=(1, N))
estimation = cumulative_variance_matrix(samples)

samples = samples.ravel()
estimation = estimation.ravel()

denom = np.arange(estimation.shape[0]) + 2
left = estimation - z * (np.sqrt(2) * estimation) / np.sqrt(denom)
right = estimation + z * (np.sqrt(2) * estimation) / np.sqrt(denom)
```

<b><font color="#00b565" size="5">Получил верный интервал </font></b>

<b><font color="#00b565" size="3">Замечание:</font></b> <i>Получен неправильный интервал.</i>

* Неудобный масштаб графика. Комм. Стоило построить либо в plotly, либо установить удобный ylim.

* Нет названия у графика. Комм. Стоило как-то назвать графики, графики потеряли немного в наглядности и информативности. 

* Нет истинного значения на графике. </i>

```python
draw_confidence_interval(
    [left],
    [right],
    [estimation],
    samples,
    ylim=(-2, 4),   # Удобный ylim
    dots_flag=True,
    title_name="Зависимость д.и. от размера выборки",
    true_values=[sigma]
)
```

## Задача 5.

Пусть случайная величина $X$ имеет нормальное распределение с параметрами $(a, \sigma^2)$, то есть:
$$
X \sim \mathcal{N}(a, \sigma^2)
$$
Рассмотрим следующую задачу оптимизации:
$$
f(X) = (\mathsf{E}X-1)^2 + (\mathsf{D}X - 1)^2 \longrightarrow \min_{a, \sigma^2}
$$
В данном случае правильный ответ мы можем легко найти непосредственно, однако в реальности возникают гораздо более сложные функции, и решить задачу напрямую руками не представляется возможным. Для решения таких задач применяются различные градиентные методы, такие как, например, стохастический градиентный спуск. Оказывается, что в нём критически важно использовать несмещенные оценки градиента.

*Замечание.* Если смещение заключается только в том, что математическое ожидание оценки отличается от вектора градиента домножением каждой компоненты на одну и ту же константу, то проблем нет &mdash; все равно мы используем шаг градиента.

В данной задаче вам предлагается на примере простой функции убедиться, насколько важным оказывается использовать несмещенные оценки в итерационных процедурах.

**Решение:**

Запишите оптимальные значения параметров $a$ и $\sigma^2$, а также шаг простого градиентного спуска для минимизаци определенной выше функции $f(X)$:

Оптимальные значения: $a = 1$ и  $\sigma^2 = 1$

GD: $a_{k+1} = a_k - \eta \frac{\partial f}{\partial(a)}(a_k) = a_k - 2\eta (a_k - 1)$

$\ \ \ \ \ \ $   $\sigma^2_{k+1} = \sigma^2_k - \frac{\partial f}{\partial(\sigma^2)}(\sigma^2_k) = \sigma^2_k - 2\eta (\sigma_k - 1)$

Реализуйте метод простого градиентного спуска для выше описанной задачи оптимизации. Для каждой итерации сохраните текущие значения среднего и дисперсии, постройте график зависимости значений $a$ и $\sigma^2$ от шага процедуры. Наблюдается ли сходимость к оптимальным параметрам?

```python
n_iter = 300
lr = 0.01
eta_1 = lr
eta_2 = lr
a = 1
sigma2 = 1
```

  * <b><font color="#00b565" size="5">Исправил метод Монте-Карло</font></b>

<b><font color="#00b565" size="3">Замечание:</font></b> Инд. комм. Монте-Карло ни в одном из случаев не сходится к нужному значению. 



```python
def MK_grad(a, sigma2, sigma_bias=0):
    samples = np.random.normal(loc=a, scale=np.sqrt(sigma2), size=5)

    grad_a = 2 * (np.mean(samples) - 1)
    grad_sigma = 2 * (samples.var(ddof=sigma_bias) - 1)

    return (grad_a, grad_sigma)
```

  * <b><font color="#00b565" size="5">Конец исправлений</font></b>


<b><font color="#00b565" size="5">Исправил график</font></b>

<b><font color="#00b565" size="3">Замечание:</font></b> <i> 

* Плохой график. Комм. Информативная часть графика слишком маленькая. Инд. комм. Точки режут глаз, возможно будет лучше выглядеть в большем масштабе

```python
def gradient_descent(
    n_iter=100,
    eta_a=0.001,
    eta_sigma2=0.001,
    a_init=0,
    sigma2_init=0,
    MK_flag=False,
    sigma_bias=0,
):
    """
    Реализация простого градиентного спуска для нахождения параметров a и sigma^2.

    :param n_iter: Количество итераций
    :param eta_a: Шаг градиентного спуска для параметра a
    :param eta_sigma2: Шаг градиентного спуска для параметра sigma^2
    :param a_init: Начальное значение a
    :param sigma2_init: Начальное значение sigma^2
    :param MK_flag: Флаг для использования альтернативной функции обновления
    :param sigma_bias_flag: Флаг для использования сдвига в параметре sigma
    :return: История изменения параметров a и sigma^2
    """
    a_s = [a_init]
    sigma2_s = [sigma2_init]

    a_new = a_init
    sigma2_new = sigma2_init

    for i in range(n_iter):
        if MK_flag:
            a_grad, sigma_grad = MK_grad(a_new, sigma2_new, sigma_bias)
            a_new -= eta_a * a_grad
            sigma2_new -= eta_sigma2 * sigma_grad
        else:
            a_new = a_s[-1] - 2 * eta_a * (a_s[-1] - 1)
            sigma2_new = sigma2_s[-1] - 2 * eta_sigma2 * (sigma2_s[-1] - 1)

        a_s.append(a_new)
        sigma2_s.append(sigma2_new)

    return a_s, sigma2_s


def plot_convergence(a_s, sigma2_s, sigma_true_val=None):
    """
    Строит графики сходимости параметров a и sigma^2.

    :param a_s: История значений параметра a
    :param sigma2_s: История значений параметра sigma^2
    """
    fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

    # График для a
    ax1.plot(np.arange(len(a_s)), a_s, label="a", color="blue", marker=" ", ls="-")
    ax1.set_title(r"Сходимость параметра $a$")
    ax1.set_xlabel("Итерация")
    ax1.axhline(y=1, color="green", linestyle="--", label="a = 1")
    ax1.set_ylabel("Значение a")
    ax1.grid(True)

    # График для sigma^2
    ax2.plot(
        np.arange(len(sigma2_s)),
        sigma2_s,
        label=r"$\sigma^2$",
        color="green",
        marker=" ",
        ls="-",
    )
    if sigma_true_val is not None:
        ax2.axhline(y=sigma_true_val, color="r", ls="--", label="Теоретическое значение оценки")
    ax2.set_title(r"Сходимость параметра $\sigma^2$")
    ax2.set_xlabel("Итерация")
    ax2.axhline(y=1, color="green", linestyle="--", label=r"$\sigma^2 = 1$")
    ax2.set_ylabel(r"Значение $\sigma^2$")
    ax2.grid(True)

    plt.tight_layout()
    plt.show()


a_s, sigma2_s = gradient_descent(
    n_iter=n_iter, eta_a=eta_1, eta_sigma2=eta_2, a_init=0, sigma2_init=0
)

plot_convergence(a_s, sigma2_s)
```

<b><font color="#00b565" size="5">Конец исправлений</font></b>

Как видим, параметры действительно сходятся к оптимальным параметрам.
